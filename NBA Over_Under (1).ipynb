```python
import pandas as pd
import numpy as np
```

```python
from nba_api.stats.endpoints import leaguestandings
```

```python
standings = leaguestandings.LeagueStandings(season="2023-24")
```

```python
data = standings.get_data_frames()[0]
```

```python
selection = data[['TeamName', 'Conference','Division','Record','WINS', 'LOSSES', 'WinPCT']] 
```

```python
projected_final = round(selection['WinPCT'] * 82)
selection['Projected wins'] = projected_final
```

```python
over_under_dict = {
'Hawks':42.5,
'Celtics': 54.5,
'Nets': 37.5, 
'Hornets': 31.5, 
'Bulls': 37.5, 
'Cavaliers': 50.5, 
'Mavericks': 43.5, 
'Pistons': 27.5, 
'Nuggets': 52.5, 
'Warriors': 47.5, 
'Rockets': 31.5,
'Pacers': 38.5,
'Clippers': 45.5,
'Lakers': 46.5,
'Grizzlies': 45.5,
'Heat': 44.5,
'Timberwolves': 44.5,
'Bucks': 53.5,
'Pelicans': 44.5,
'Knicks': 45.5,
'Thunder': 44.5,
'Magic': 37.5,
'Suns': 52.5,
'76ers': 47.5,
'Trail Blazers': 27.5,
'Kings': 44.5,
'Spurs': 28.5,
'Raptors': 36.5,
'Jazz': 35.5,
'Wizards': 23.5,
    }
```

```python
selection['Over/Under'] = selection['TeamName'].map(over_under_dict)
```

```python
selection['Projection'] = np.where(selection['Projected wins'] > selection['Over/Under'], 'Over', 'Under')
```

```python
def find_right(row):
#     red = 'background-color: red;'
#     green = 'background-color: green;'
#     default = ""

    key = row.loc["Projection"]
    scott = row.loc["Scott"]
    o = row.loc["Olver"]
    
    if key == o:
         color =  'green'
    else:
         color =  'red'
 
    return ['background-color: {}'.format(color) for i in row]

mycolor = mydf.style.apply(find_right, axis=1 )
```

```python
df = mydf

# Custom function to apply color based on matching values
def color_matching(s):
    # Create an empty DataFrame with the same index and columns as the original
    color_df = pd.DataFrame('', index=s.index, columns=s.columns)
    
    # Iterate over each row
    for i, row in s.iterrows():
        # Iterate over 'Scott', 'Oliver', 'Miles', 'Carolyn' columns
        for col in ['Scott', 'Oliver', 'Miles', 'Carolyn']:
            # Compare the value with the value in 'Projection' column
            if row[col] == row['Projection']:
                # If values match, set the background color to green
                color_df.at[i, col] = 'background-color: green'
            else:
                # If values don't match, set the background color to red
                color_df.at[i, col] = 'background-color: red'
    
    return color_df

# Apply the custom function to color the DataFrame
styled_df = df.style.apply(color_matching, axis=None)

# Display the styled DataFrame
styled_df
```

```python
# Initialize counters for each name
matches = {'Scott': 0, 'Oliver': 0, 'Miles': 0, 'Carolyn': 0}

# Iterate over each row
for _, row in df.iterrows():
    # Iterate over each name
    for name in ['Scott', 'Oliver', 'Miles', 'Carolyn']:
        # Check if the value in the name column matches the value in the 'Projection' column
        if row[name] == row['Projection']:
            # Increment the counter for the matching name
            matches[name] += 1

# Print the counts
for name, count in matches.items():
    print(f"{name}: {count} matches")
```

```python
matches_df = pd.DataFrame(list(matches.items()), columns=['Name', 'Matches'])

# Display the DataFrame
matches_df.style.hide_index()
```

```python
def highlight(row):
    red = 'background-color: red;'
    green = 'background-color: green;'
    default = ""
    
    if row["Projection"]  == row["Scott"]:
        return [default, green]
    elif:
        return [default, red]
    else:
        return [default, red]
    
    
color = mydf.style.apply(highlight, subset=["Projection", "Scott"], axis=1 )
```
